```python
# answer 21 // dataframe employee
import pandas as pd

# Create Employee DataFrame
Employee = pd.DataFrame({
    'Name': ['Atul','Raj','Darpan','Anmol','Piyush'],
    'Age': [25,30,23,22,32],
    'Designation': ['Manager','Clerk','Analyst','Clerk','Manager'],
    'Salary': [56000,25000,35000,28000,58000],
    'Bonus': [15000,7000,9000,13000,12000]
}, index=['E101','E102','E103','E104','E105'])

print("Initial Employee DataFrame:")
print(Employee)

# a) Add a new row
Employee.loc['E106'] = ['GEET', 0, 'Analyst', 70000, 15000]
print("\nAfter adding new employee:")
print(Employee)

# b) Add column Shift
Employee['Shift'] = 'Morning'
print("\nAfter adding Shift column:")
print(Employee)

# c) Youngest employee
print("\nYoungest employee:")
print(Employee[Employee['Age'] == Employee['Age'].min()])

# d) Employees with Designation Clerk
print("\nEmployees with designation Clerk:")
print(Employee[Employee['Designation'] == 'Clerk'])

# e) Average Bonus
print("\nAverage Bonus:")
print(Employee['Bonus'].mean())

# f) Total Salary
print("\nTotal Salary:")
print(Employee['Salary'].sum())

# g) Add City column
Employee['City'] = [
    'Janak Puri','Tilak Nagar','Shiv Nagar',
    'Vikas Puri','Janak Puri','Tilak Nagar'
]
print("\nAfter adding City column:")
print(Employee)

# h) Delete employees E103 and E105
Employee.drop(['E103','E105'], inplace=True)
print("\nAfter deleting employees E103 and E105:")
print(Employee)

# i) Rename Bonus column to Incentive
Employee.rename(columns={'Bonus':'Incentive'}, inplace=True)
print("\nAfter renaming Bonus to Incentive:")
print(Employee)
```

```python
#ANSWER 22
import pandas as pd
import numpy as np

# =======================
# PART 1 : Items DataFrame
# =======================

# Step 0: Create the DataFrame
data = pd.DataFrame({
    'Count': [15, 20, 25, 35],
    'Price': [250, 200, 600, 900]
}, index=['Pencil', 'Pen', 'Eraser', 'Scale'])

print("Initial DataFrame:")
print(data)

# a) Change the Eraser count to 15
data.loc['Eraser', 'Count'] = 15
print("\nAfter changing Eraser count to 15:")
print(data)

# b) Rename index “Scale” as “Wooden Scale”
data.rename(index={'Scale': 'Wooden Scale'}, inplace=True)
print("\nAfter renaming Scale to Wooden Scale:")
print(data)

# c) Change all row labels
data.index = ['Product 1', 'Product 2', 'Product 3', 'Product 4']
print("\nAfter changing all row labels:")
print(data)

# d) Change column names to “Discount”, “Total”
data.columns = ['Discount', 'Total']
print("\nAfter changing column names:")
print(data)

# e) Add new row “Product 5” with values 10 and 100
data.loc['Product 5'] = [10, 100]
print("\nAfter adding Product 5:")
print(data)

# f) Create new DataFrame “Products” without column “Total”
Products = data.drop(columns=['Total'])
print("\nNew DataFrame Products (without Total column):")
print(Products)


# ==================================
# PART 2 : Agricultural Production DF
# ==================================

# Step 0: Create the DataFrame
df = pd.DataFrame({
    'Fruits': [44, 11950, 7152, 140169, 7830, 113],
    'Pulses': [23, 818, 33, 3184, 931, 1],
    'Rice': [814, 1930, 11586, 13574, 7452, 2604],
    'Wheat': [5, 2737, 16440, 30056, np.nan, np.nan]
}, index=['Tripura', 'Gujrat', 'Punjab', 'UP', 'AP', 'Kerala'])

print("\nAgricultural Production DataFrame:")
print(df)

# a) Total production for each state
print("\nTotal production for each state:")
print(df.sum(axis=1))

# b) Total production for each item type
print("\nTotal production for each item type:")
print(df.sum())

# c) How many states produce Wheat?
print("\nNumber of states producing Wheat:")
print(df['Wheat'].count())
```
